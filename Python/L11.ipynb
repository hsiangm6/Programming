```python
#1
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
x=np.linspace(0, np.pi*4, 51)
plt.plot(x, np.sin(x), c='dodgerblue', linewidth=4.0)
plt.plot(x, np.cos(x), c='g', linewidth=4.0)
plt.xlim(-0.5, 13)
plt.ylim(-2, 2)
plt.ylabel('hi there', size=20)
plt.legend(['sin', 'cos'])
plt.title('sin cos Plot Lab', c='r', size=30)
```

```python
#3
# Fixing random state for reproducibility
np.random.seed(19680801)

dt = 0.01
t = np.arange(0, 30, dt)
nse1 = np.random.randn(len(t))                 # white noise 1
nse2 = np.random.randn(len(t))                 # white noise 2

# Two signals with a coherent part at 10Hz and a random part
s1 = np.sin(2 * np.pi * 10 * t) + nse1
s2 = np.sin(2 * np.pi * 10 * t) + nse2

fig, axs = plt.subplots(2, 1)
axs[0].plot(t, s1, linewidth=1)
axs[0].plot(t, s2, c='darkorange', linewidth=1)
axs[0].set_xlim(0, 2)
axs[0].set_ylim(-4, 4)
axs[0].set_xlabel('time')
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)

cxy, f = axs[1].cohere(s1, s2, 256, 1. / dt, linewidth=1)
axs[1].set_ylim(-0.1, 1.1)
axs[1].set_ylabel('coherence')

fig.tight_layout()
plt.show()
```
